```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pnt
import seaborn as sns

data = pd.read_csv("//home//primes//Documents//COOPERATIVE UNIVERSITY OF KENYA//4.1//MACHINE-LEARNING//bestsellers with categories.csv")
                   
data.head()
```

```python
pnt.title('Price Distribution')
sns.distplot(data['Price'])
pnt.show()
```

```python
sns.histplot(data['User Rating'], bins=10, kde=True)
pnt.title('Distribution of User Ratings')
pnt.show()
```

```python
sns.histplot(data['Reviews'], bins=20, kde=True)
pnt.title('Distribution of Reviews')
pnt.show()
```

```python
sns.scatterplot(x='Price', y='User Rating', data=data)
pnt.title('Price vs User Rating')
pnt.show()

sns.scatterplot(x='Reviews', y='User Rating', data=data)
pnt.title('Reviews vs User Rating')
pnt.show()
```
